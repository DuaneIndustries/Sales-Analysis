### Average Tip Amount (Tipped Transactions)

```python
df['Tip'] = df['Tip'].str.replace('\$', '')
df['Tip'] = df['Tip'].astype(float)
tiplist = df.sort_values(['Tip'], ascending=False)
```

```python

plt.figure(figsize=(8,5))
bins = [0.25, 0.50, 0.75, 1.00, 1.25, 1.50, 1.75, 2.00, 2.25, 2.50, 2.75, 3.00]

plt.hist(df.Tip, bins = bins, edgecolor='w')
plt.title('Tip Amounts and Their Frequency')
plt.xticks(bins)
plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
plt.show()
```

> ##### I've excluded transactions without a tip from this amount, the frequency for that would be 1391, as seen above.
> ##### I've capped this at 3.00 to keep the graph from becoming too stretched, but you can see tip amounts around the dollar tend to be significantly higher.  

### Average Tip Percent

```python
df['Gross Sales'] = df['Gross Sales'].str.replace('\$', '')
df['Gross Sales'] = df['Gross Sales'].astype(float)
```

```python
df['Tip Percent'] = df['Tip'] / df['Gross Sales']
```

```python
df['Tip Percent']
df.head(5)
```

```python
temp_df = df[df.Tip != 0.00]
temp_df.head()
```
